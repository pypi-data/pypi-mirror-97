```python
import fitsio
import matplotlib.pyplot as plt
import numpy as np
import time
import coord
%matplotlib inline
```

```python
data = fitsio.read('/Users/Mike/Downloads/sva1_gold_r1.0_catalog.fits')
```

```python
print(data.dtype.names)
```

```python
mask = (data['RA'] > 50) & (data['RA'] < 100) & (data['DEC'] < -40) & (data['DEC'] > -65)
data = data[mask]
```

```python
print(data.shape)
N = 1000000
indx = np.random.choice(len(data), size=N, replace=False)
ra = data['RA'][indx]
dec = data['DEC'][indx]
X = np.array([ra, dec]).T
print(X.shape)
```

```python
ra_rad = ra * (coord.degrees / coord.radians)
dec_rad = dec * (coord.degrees / coord.radians)
x, y, z = coord.CelestialCoord.radec_to_xyz(ra_rad, dec_rad)
X2 = np.array([x, y, z]).T
```

```python
import matplotlib.cm as cm
npatch = 40
colors = cm.rainbow(np.linspace(0, 1, npatch))
np.random.shuffle(colors)
```

```python
import kmeans_radec
t0 = time.time()
km = kmeans_radec.kmeans_sample(X, npatch, maxiter=100, tol=1.0e-5)
t1 = time.time()
print(t1-t0)
```

```python
plt.scatter(ra, dec, color=colors[km.labels], s=0.1)
plt.show()
```

```python
import scipy
from scipy.cluster.vq import kmeans2
t2 = time.time()
cen, labels =kmeans2(X2,npatch,iter=100,minit='points')
t3 = time.time()
print(t3-t2)
```

```python
plt.scatter(ra, dec, color=colors[labels],s=0.1)
plt.show()
```

```python
import pyclustering.cluster.kmeans
import pyclustering.cluster.center_initializer
# create instance of K-Means algorithm
t4 = time.time()
py_centers = pyclustering.cluster.center_initializer.kmeans_plusplus_initializer(X2, npatch).initialize();
kmeans_instance = pyclustering.cluster.kmeans.kmeans(X2, py_centers, ccore=False)
# run cluster analysis and obtain results
kmeans_instance.process()
py_clusters = kmeans_instance.get_clusters()
t5 = time.time()
print(t5-t4)
```

```python
py_labels = np.zeros_like(ra, dtype=int)
for i, cl in enumerate(py_clusters):
    py_labels[cl] = i
plt.scatter(ra, dec, color=colors[py_labels], s=0.1)
plt.show()
```

```python
import sklearn.cluster
t6=time.time()
sk_kmeans = sklearn.cluster.KMeans(n_clusters=npatch, random_state=0, n_init=1, max_iter=100)
sk_kmeans.fit(X2)
t7=time.time()
print(t7-t6)
```

```python
plt.scatter(ra, dec, color=colors[sk_kmeans.labels_], s=0.1)
plt.show()
```

```python
import treecorr
t8 = time.time()
cat = treecorr.Catalog(ra=ra, dec=dec, ra_units='deg', dec_units='deg')
t9 = time.time()
print(t9-t8)
cat.write('des_sv.fits')
```
