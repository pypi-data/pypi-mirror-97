```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import label_ranking_average_precision_score

def map_scores(train_set, true_ssm):
    tfidf_vectorizer = TfidfVectorizer()
    tfidf_matrix_train = tfidf_vectorizer.fit_transform(train_set)
    scores = cosine_similarity(tfidf_matrix_train, tfidf_matrix_train)
    return label_ranking_average_precision_score(true_ssm, scores), scores

def plot_ssm(M, axes, axs):    
    axes[axs].matshow(M, cmap='jet')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 5))

map_score, scores = map_scores(docs_s, true_ssm)
plot_ssm(scores, axes, 0)
print(map_score)

map_score, scores = map_scores([' '.join(np.unique(dc.split(' '))) for dc in docs_s], true_ssm)
plot_ssm(scores, axes, 1)
print(map_score)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 5))

map_score, scores = map_scores(docs_ct, true_ssm)
plot_ssm(scores, axes, 0)
print(map_score)

unique_ct = [' '.join(np.unique(dc.split(' '))) for dc in docs_ct]
map_score, scores = map_scores(unique_ct, true_ssm)
plot_ssm(scores, axes, 1)
print(map_score)

#[' '.join(np.unique(dc.split(' '))) for dc in docs]

plt.show()
```

```python
def ngrams(arr_tokens, n):
    ngrams = ['$'.join(x) for x in zip(*[arr_tokens[i:] for i in range(n)])]
    return ' '.join(ngrams)
```

```python
docs_ct_2gram = [ngrams(n.split(' '), 2) for n in docs_s]
docs_ct_3gram = [ngrams(n.split(' '), 4) for n in docs_s]
docs_ct_4gram = [ngrams(n.split(' '), 6) for n in docs_s]
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(16, 5))

map_score, scores = map_scores(docs_ct_2gram, true_ssm)
plot_ssm(scores, axes, 0)
print(map_score)

map_score, scores = map_scores(docs_ct_3gram, true_ssm)
plot_ssm(scores, axes, 1)
print(map_score)

map_score, scores = map_scores(docs_ct_4gram, true_ssm)
plot_ssm(scores, axes, 2)
print(map_score)



#[' '.join(np.unique(dc.split(' '))) for dc in docs]

plt.show()
```
