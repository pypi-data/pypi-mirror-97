# KiWIS Basic Example
In this notebook we show how to create a connection to KiWIS and access basic data

```python
from kisters.water.time_series.kiwis import KiWIS
```

```python
# Create kiwis connection
kiwis = KiWIS(kiwis_uri="http://kiwis.kisters.de/KiWIS2/KiWIS", datasource=0, create_pandas=True)
```

```python
kiwis.getStationList()
```

```python
kiwis.getTimeseriesList(station_id=23438)
```

```python
df = kiwis.getTimeseriesValues(ts_id=7411042, from_='2018-01-01', to='2018-01-02')
```
